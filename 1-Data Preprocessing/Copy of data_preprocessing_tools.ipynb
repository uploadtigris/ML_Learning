# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np #work with arrays
import matplotlib.pyplot as plt #chart and stuff
import pandas as pd #import dataset, matrix of features, dependent variable vector
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv') #library.function('argument')
#matrix of features (columns)
X = dataset.iloc[:, :-1].values #iloc[rows, col_range] = locate indexes; : = range... all the rows; :-1 = index of the last column (excludes the last column)
#dependent variable vector (last column)
Y = dataset.iloc[:, -1].values
```

```
print(X) #features
```

```
print(Y) #dependent variable vector
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3]) #only specify columns WITHOUT strings
X[:, 1:3] = imputer.transform(X[:, 1:3]) #X[rows, columns]
```

```
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

```
#encode country column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### Encoding the Dependent Variable

```
#encode last column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```
print(Y)
```

## Splitting the dataset into the Training set and Test set

```
#we must split BEFORE because we need to find the Sigma of only the training set data.
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(Y_train)
```

```
print(Y_test)
```

## Feature Scaling

```
#Standardisation will work ALL of the time
#Normalisation works best when the data is normally distributed
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#fit will only grab the mean and standard deviation
#transform will compute the entire X standardisation
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```

```
print(X_train)
```

```
print(X_test)
```
